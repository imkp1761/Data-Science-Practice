```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('sugarlevel')
plt.ylabel('no of patients')
plt.title('sugar report')
plt.grid()
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_men,blood_sugar_women] , bins=[70,100,130,150], rwidth=0.95,color=['red','green'] , 
         label=['men','women'])
plt.legend()
```
