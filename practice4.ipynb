```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7,8]
y=[51,12,63,14,85,16,97,18]
```

```python
plt.plot(x,y,color='green')
```

```python
plt.xlabel('day')
plt.ylabel('temp')
plt.title('weather')
plt.plot(x,y,'g+')
```

```python
days = [1,2,3,4,5,6,7]
max_t= [50, 51, 52,48,47,49,46]
min_t= [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,color='green')
plt.plot(days,min_t,color='blue')
plt.plot(days,avg_t,color='red')
```

```python
plt.xlabel('day')
plt.ylabel('temp')
plt.title('weather')
plt.plot(days,max_t,color='green',label='max')
plt.plot(days,min_t,color='blue',label='min')
plt.plot(days,avg_t,color='red',label='avg')

plt.legend()
```
